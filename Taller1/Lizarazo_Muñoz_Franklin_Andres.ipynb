```python
##5 Programa que recibe una cadena de texto y calcula cuantas:
    ##a Mayusculas tiene
    ##b Cuantas letras con tilde tiene
    ##c Cuantos digitos tiene
    ##d Cuantos espacios tiene
    ##e Cuantas palabras reservadas tiene

import keyword

##Condiciones iniciales
i=0
up=0
tild=0
digit=0
space=0
res=0
keys=keyword.kwlist ##Lista de palabras reservadas
tildes=[225,193,233,201,237,205,243,211,250,218]
    
##Entrada de datos
string=input('Introduzca un texto: ')

##Conteo
for n in string:
    o=ord(n)
    
    if n.isupper(): ##Cuenta mayusculas
        up+=1
        
    if n.isdigit(): ##Cuenta digitos
        digit+=1
        
    if n.isspace(): ##Cuenta espacios
        space+=1
        
    if o in tildes: ##Busca letras con tilde
        tild+=1
        
while i<len(keys): ##Busca palabras reservadas
    if keys[i] in string:
        res+=1
    i+=1

        
##Salida de datos
print('\nEl numero de mayusculas es:',up,'\nEl numero de digitos es: ',digit,
      '\nEl numero de espacios es: ',space,'\nEl numero de letras con tildes es: ',tild,
      '\nEl numero de palabras reservadas dentro del texto es: ',res)
```

```python
##6 Leer una cadena de texto y organizarla alfabeticamente

##Condiciones iniciales
i=0
execution=True ##Mantiene la ejecucion
order=[]
digits=['1','2','3','4','5','6','7','8','9','0']

##Entrada de datos
string=input('Introduzca un texto: ').lower()

##Busca digitos dentro del texto para generar una excepcion
while i<len(digits):
    if digits[i] in string:
        execution=False
    i+=1
    
##Mi version del try except
if execution:
    for n in string:
        if n.isspace()==False:
            order.append(n)
    order.sort()
    
    ##Salida de datos
    print('\nEl texto ordenado alfabeticamente (en minusculas) sin espacios es',''.join(order))
else:
    ##Except
    print('\nHa ocurrido un error, No introduzca numeros')
    
```

```python
##7 Leer una lista de números verificar que está ordenada. Luego insertar un numero donde le corresponde

##Condiciones iniciales
order=[]
ordered=True

##Funcion para convertir los input a int o float
def float_or_int(x):
    if '.' in x:
        y=float(x)
    else:
        y=int(x)
        
    return y

##Try por si ingresan letras
try:
    ##Entrada de datos
    n=int(input('Digite la cantidad de numeros que va a ingresar: '))
    for x in range(0,n):
        o=float_or_int(input(f'Digite el numero {x+1} de la lista: '))
        order.append(o)
    
    num=float_or_int(input('Digite el numero que desea insertar en la lista: '))


    ##Ciclo que comprueba que la lista este ordenada
    for i in range (0,len(order)):
        if i!=0:
            if order[i]<order[i-1] or order[i]==order[i-1]:
                ordered=False
        
        ##Insercion del numero
        if num<order[0]:
            order.insert(0,num)
        elif num>order[len(order)-1]:
            order.insert(len(order),num)
        elif num>order[i-1] and num<order[i]:
            order.insert(i,num)

    ##Salida de resultados
    if ordered:
        print('\nLa lista esta ordenada y con el numero insertado queda igual a:\n',order)
    else:
        print('\nLa lista no esta ordenada, por lo que no se puede insertar el numero')
    
except:
    print('\nHa ocurrido un error, los datos ingresados no son validos')
```

```python
##8 Leer una lista de números enteros y determinar cual es el segundo elemento mayor de dicha lista.

##Condiciones iniciales
List=[]
new_list=[]

##Funcion para convertir los input a int o float
def float_or_int(x):
    if '.' in x:
        y=float(x)
    else:
        y=int(x)
        
    return y
    
##Try por si ingresan letras
try:
    ##Entrada de datos
    n=int(input('Cuantos numeros tendra su lista?: '))
    
    for x in range(0,n):
        o=float_or_int(input(f'Digite el numero {x+1} de su lista: '))
        List.append(o)
        
        
    ##Encuentra el valor maximo
    max_1=List[0]

    for x in range(0,len(List)):
        if List[x]>max_1:
            max_1=List[x]
    
    ##Crea una nueva lista sin numeros repetidos y sin el valor maximo
    for k in List:
        if k!=max_1:
            if k not in new_list:
                new_list.append(k)
                
    ##Comprueba que no todos sean el mismo numero
    if len(new_list)==0:
        
        print('\nLa lista no tienen ningun numero mayor que otro ya que todos son el mismo valor')
    else:
        
        ##Halla el valor maximo de la nueva lista
        max_2=new_list[0]
    
        for i in range(0,len(new_list)):
            if new_list[i]>max_2:
                max_2=new_list[i]
        
        ##Salida de Resultados
        print('\nEl segundo numero mayor de la lista es: ',max_2)
except:
    print('\nHa ocurrido un error, los datos ingresados no son validos')
```

```python
##10 Leer una matriz e imprimir el vector que resulta de desenredar una matriz en forma de caracol

##Try por si en las filas y columnas no digitan numeros enteros
try:
    
    ##Numero de filas y columnas
    n=int(input('Digite el numero de filas de la matriz: '))
    m=int(input('Digite el numero de columnas de la matriz: '))
    
    #Condiciones iniciales
    A=[] #Matriz
    a=[] #Vectores de la matriz
    v=[] #Vector resultante
    
    x=n #Posicion en x de las iteraciones
    y=m #Posicion en y de las iteraciones

    n_max=n-1 #Fila maxima que puede haber en la matriz
    m_max=m-1 #Columna maxima que puede haber en la matriz

    n_min=0 #Fila minima de la matriz
    m_min=0 #Fila maxima de la matriz

    la=n*m #Numero de items de la matriz
    lv=len(v) #Numero de items del vector
    
    ##Contadores que hacen que se desenrosque en caracol
    count=1
    minus=0

    x-=1
    y-=1

    ##funcion que determina si un item es int, float o str y lo convierte
    ##teniendo en cuenta que no se realizan operaciones decidi admitir letras como items dentro de la matiz
    def int_float_string(item):
        try:
            if '.' in item:
                return float(item)
            else:
                return int(item)
        except:
            return item

    ##Llenado de la matriz
    for k in range(0,n):
        for j in range(0,m):
            it=int_float_string(input(f'Digite el item [{k+1}][{j+1}]: '))
            a.append(it)
        print('\n')
        A.append(a)
        a=[]
        
    ##Ciclo para que las ordenes se ejecuten hasta que el vector tenga el mismo contenido que la matriz
    while lv<la:
        
        ##Comprueba si las iteraciones deben ser de derecha a izquierda o viceversa
        if x==n_max and y==m_max:
            for i in range (0,m_max-minus):
                y=m_max-i
                last_m=A[x][y]
                v.append(last_m)
            y-=1
        elif x==n_min and y==m_min:
            for i in range (0,m_max-minus):
                y=m_min+i
                last_m=A[x][y]
                v.append(last_m)
            y+=1
            
        ##Caso especial para numero de filas impar
        if n%2!=0 and lv==la-1:
            y=m_max
            if m>n:
                x+=1
                y-=1
            last_m=A[x][y]
            v.append(last_m)
        
            break
        
        ##Comprueba si las iteraciones deben ser hacia arriba o hacia abajo
        if x==n_max and y==m_min:
            for i in range (0,n_max-minus):
                x=n_max-i
                last_n=A[x][y]
                v.append(last_n)
            x-=1
        if x==n_min and y==m_max:
            for i in range (0,n_max-minus):
                x=n_min+i
                last_n=A[x][y]
                v.append(last_n)
        
        ##Reduce los parametros de la matriz cada dos ejecuciones
        if count%2==0:
            n_max-=1
            m_max-=1
    
            n_min+=1
            m_min+=1
        
            minus+=1
            y-=1
            
        ##Actualiza datos tras cada ejecucion
        count+=1
        lv=len(v)
    
    print('El vector resultante de desenroscar la matriz desde la esquina infeior derecha es: \n',v)
    
except:
    
    print('Ha ocurrido un error, asegurese de que los datos ingresados son validos')
```

```python
##11 Diseñar una funcion que pida alto, ancho y el caracter para dibujar un rectangulo

##Funcion
def draw_rectangule():
    '''funcion que pide ancho, alto y el caracter para dibujar un rectangulo'''
    
    ##Try en caso de que se digite float o str
    try:
        
        ##Convertir a unidades necesarias, en caso de que no vengan asi por defecto
        wi=int(input('Digite el ancho del rectangulo: '))
        he=int(input('Digite la altura del rectangulo: '))
        ch=input('Digite el caracter con el que desea que se dibuje el rectangulo: ')
        
        print('\n')##Espacio antes del dibujo (por presentacion)
        
        ##Ciclo padre para las filas
        for i in range(0,he):
            
            ##Ciclo hijo para las columnas
            for i in range(0,wi):
                print(ch,end='')
            print(' ')
            
    except:
        print('Ha ocurrido un error, los valores de alto y ancho deben ser numeros enteros')
        
##Prueba de la funcion
draw_rectangule()
```

```python
##12 Diseñar una funcion que pida la anchura de un triangulo y lo dibuje con *

##Funcion
def draw_triangule():
    '''Funcion que pide el ancho de un triangulo y lo dibuja con *'''
    
    ##Try en caso de que el ancho no sea un numero entero
    try:
        ##Conversion de parametros en caso de que no vengan asi por defecto
        wi=int(input('Digite el ancho del triangulo: '))
        print('\n')##Espacio antes del dibujo (por presentacion)
        
        for i in range(0,wi):
            
            ##Ciclo hijo para las columnas
            for j in range(0,i+1):
                print('*',end='')
            print('')
        
        for n in range(-wi,0):
            
            ##Ciclo hijo para las columnas
            for k in range(n+1,0):
                print('*',end='')
            print('')
            
            
    except:
        print('Ha ocurrido un error, los valores de alto y ancho deben ser numeros enteros')


##Prueba de funcion
draw_triangule()
        
```

```python
##13 Diseñar una funcion que pida dos años y diga cuantos años biciestos hay entre ambos, incluyendolos

def años_biciestos():
    '''Funcion que calcula el numero de años biciestos entre dos años que se introduzcan'''
    
    ##Se importa la libreria por si no se tenia antes
    import numpy as np
    
    ##Condiciones iniciales
    i=0
    count=0
    
    ##Try en caso de que los valores no sean numeros
    try:
        ##Conversion de los parametros en caso sea necesario
        y1=year1=int(input('Digite el año inicial: '))
        y2=year2=int(input('Digite el año final: '))
        years=np.arange(y1,y2+1)
        
        ##Ciclo que cuenta los años
        while i<len(years):
            if years[i]%4==0:
                count+=1
            i+=1

        ##Salida de resultados
        print('\nEl numero de años biciestos entre ',y1,' y ',y2,' es: ',count)
        
    except:
        print('\nHa ocurrido un error, los valores digitados no son validos')

##Prueba de la funcion
años_biciestos()
```

```python
##14 Diseñar una funcion que proponga sumas entre dos numeros positivos entre 1 y 100 y comprueve 5 aciertos

##Funcion
def sumas_random():
    '''Funcion que propone sumas aleatorias y comprueba los resultados, hasta que se acierten 5'''
    
    ##Importa la libreria por si no se tiene
    import random
    
    ##Condiciones iniciales
    hits=0
    num_min=1
    num_max=100
    
    ##En caso de que digiten letras
    try:
        ##Que se ejecute hasta que acierte 5 veces
        while hits<5:
            ##Numeros aleatorios
            num1=random.randint(num_min,num_max)
            num2=random.randint(num_min,num_max)
                
            ##Ingresar respuestas
            answer=int(input(f'Cuanto es: {num1} + {num2}= '))
            
            ##Comprobacion respuesta
            if answer==num1+num2:
                hits+=1
                print('Correcto, has acertado ',hits,' respuesta(s)\n')
            else:
                print('Respuesta Incorrecta, intentalo de nuevo\n')
        
        ##Fin del programa
        print('\nFelicidades, Ganaste')
        
    except:
        print('\nHa ocurrido un error, los valores digitados no son validos')
        
sumas_random()
        
```
